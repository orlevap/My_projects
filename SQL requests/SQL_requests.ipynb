### –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–Ω–∏–≥ –≤—ã—à–µ–¥—à–∏—Ö –ø–æ—Å–ª–µ 1 —è–Ω–≤–∞—Ä—è 2000 –≥–æ–¥–∞<a class="anchor" id="4-bullet"></a>

```python
query = '''
           SELECT
               COUNT(*) AS cnt 
           FROM 
               books
           WHERE 
               publication_date >= '2000-01-01';
        '''
 
req_1 = pd.io.sql.read_sql(query, con = engine)
req_1
```

###### –í—ã–≤–æ–¥

–ü–æ—Å–ª–µ 1 —è–Ω–≤–∞—Ä—è 2000 –≥–æ–¥–∞ –≤—ã—à–ª–∞ 821 –∫–Ω–∏–≥–∞.

### –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –¥–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏<a class="anchor" id="5-bullet"></a>
üëà[–Ω–∞–∑–∞–¥ –∫ –æ–≥–ª–∞–≤–ª–µ–Ω–∏—é](#0-bullet)

```python
query = '''
           SELECT
               books.book_id,
               title,
               AVG(rating) AS avg_rating,
               sub.count_reviews
           FROM
               books
           INNER JOIN ratings ON ratings.book_id = books.book_id
            
           INNER JOIN
                    (SELECT
                         book_id,
                         COUNT(text) AS count_reviews
                     FROM
                         reviews
                     GROUP BY
                         book_id
                         
                         ) AS Sub             
                      
           ON Sub.book_id = books.book_id
           
           GROUP BY 
               books.book_id,
               books.title,
               sub.count_reviews
           ORDER BY
               book_id ASC;

        '''
 
req_2_1 = pd.io.sql.read_sql(query, con = engine) 
req_2_1.head()
```

###### –í—ã–≤–æ–¥

`avg_rating` - —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞, `cnt` - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤.

### –ò–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã–ø—É—Å—Ç–∏–ª–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–Ω–∏–≥ —Ç–æ–ª—â–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü<a class="anchor" id="6-bullet"></a>
üëà[–Ω–∞–∑–∞–¥ –∫ –æ–≥–ª–∞–≤–ª–µ–Ω–∏—é](#0-bullet)

```python
query = '''
           SELECT
               books.publisher_id AS publisher_id,
               publishers.publisher AS publisher,
               COUNT(book_id) AS cnt               
           FROM 
               books
           INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
    
           WHERE 
               num_pages >= 50
           GROUP BY
               books.publisher_id,
               publishers.publisher           
           ORDER BY
               cnt DESC
           LIMIT 5;
           
        '''
 
req_3 = pd.io.sql.read_sql(query, con = engine)
req_3
```

###### –í—ã–≤–æ–¥

–ò–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ Penguin Books –≤—ã–ø—É—Å—Ç–∏–ª–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–Ω–∏–≥ —Ç–æ–ª—â–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü - 42 —à—Ç.

### –ê–≤—Ç–æ—Ä —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥. –ö–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏<a class="anchor" id="7-bullet"></a>
üëà[–Ω–∞–∑–∞–¥ –∫ –æ–≥–ª–∞–≤–ª–µ–Ω–∏—é](#0-bullet)

```python
query = '''
           SELECT
               books.author_id AS author_id,
               authors.author AS author,
               AVG(SUBQ.AVG_rating) AS avg_by_author
           FROM 
               books
           INNER JOIN authors ON authors.author_id = books.author_id
           INNER JOIN
               (
                SELECT
                    book_id,
                    AVG_rating
                FROM
                    (SELECT
                         book_id,
                         AVG(rating) AS AVG_rating,
                         COUNT(rating) as count_rating
                     FROM 
                         ratings
                     GROUP BY
                         book_id) AS Sub 
                WHERE
                    SUB.count_rating > 50    
               
                GROUP BY
                    Sub.book_id,
                    AVG_rating


               ) AS SUBQ
           ON 
               SUBQ.book_id = books.book_id
           
           
           GROUP BY
               books.author_id,
               authors.author
           ORDER BY
               avg_by_author DESC;
           
        '''
 
req_4_1 = pd.io.sql.read_sql(query, con = engine)
req_4_1
```

###### –í—ã–≤–æ–¥

–ê–≤—Ç–æ—Ä —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥ - J.K. Rowling/Mary GrandPr√©. C—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –∞–≤—Ç–æ—Ä–∞ –ø–æ –≤—Å–µ–º –µ–≥–æ –∫–Ω–∏–≥–∞–º - 4,28.

### –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª—å—à–µ 50 –æ—Ü–µ–Ω–æ–∫<a class="anchor" id="8-bullet"></a>
üëà[–Ω–∞–∑–∞–¥ –∫ –æ–≥–ª–∞–≤–ª–µ–Ω–∏—é](#0-bullet)

```python
query = '''
           SELECT
               AVG(Subq1.count_reviews) AS avg_reviews
           FROM 
           
               (SELECT
                    username,
                    COUNT(reviews.review_id) as count_reviews
                FROM 
                    reviews
                GROUP BY
                    reviews.username
                ) AS Subq1
           INNER JOIN
               (
                SELECT
                    username,
                    COUNT(ratings.rating) AS cnt
                FROM
                    ratings 
                GROUP BY
                    ratings.username

               ) AS SUBQ2
           ON 
               SUBQ2.username = Subq1.username
           WHERE
               SUBQ2.cnt > 50;
           
        '''
 
req_5_1 = pd.io.sql.read_sql(query, con = engine)
req_5_1
```
